# 05. Check the spending habits

```python
# Check the spending habits of three tpes of customers - check product prices for each loyalty category
orders_products_merged.groupby('loyalty_flag').agg({'prices':['mean','min','max']})
```

# 06. Check the types of spenders

```python
# Target customers based on spending, create a column of mean price by users

orders_products_merged['mean_order'] = orders_products_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
orders_products_merged.shape
```

```python
orders_products_merged.head(10)
```

```python
# Categories spending flag based on mean price
orders_products_merged.loc[orders_products_merged['mean_order']<10, 'spending_flag'] = 'Low spender'
```

```python
orders_products_merged.loc[orders_products_merged['mean_order']>=10, 'spending_flag'] = 'High spender'
```

```python
orders_products_merged['spending_flag'].value_counts(dropna=False)
```

```python
orders_products_merged[['user_id', 'mean_order', 'spending_flag']].head(100)
```

# 07. Check the frequent vs non-frequent customers

```python
# Create a frequency vs non-frequency customers based on median days since previous order, create a column of median days since previous orders by users

orders_products_merged['median_days_since_last_order'] = orders_products_merged.groupby(['user_id'])['days_since_last_order'].transform(np.mean)
```

```python
orders_products_merged.shape
```

```python
orders_products_merged.head(10)
```

```python
# Categories frequent vs unfrequent users based on median days since last orders
orders_products_merged.loc[orders_products_merged['median_days_since_last_order']>20, 'Frequency_flag'] = 'Non-Frequent customer'
```

```python
orders_products_merged.loc[(orders_products_merged['median_days_since_last_order']>10) & (orders_products_merged['median_days_since_last_order']<=20), 'Frequency_flag'] = 'Regular customer'
```

```python
orders_products_merged.loc[orders_products_merged['median_days_since_last_order']<=10, 'Frequency_flag'] = 'Frequent customer'
```

```python
orders_products_merged.shape
```

```python
orders_products_merged['Frequency_flag'].value_counts(dropna = False)
```

```python
orders_products_merged[['user_id', 'median_days_since_last_order', 'Frequency_flag']].head(100)
```
